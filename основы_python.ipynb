# Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def result_vowels(text):
    return sum(1 for char in text.lower() if char in 'aeiou')

def test_result_vowels():
    assert result_vowels("Hello World") == 3
    assert result_vowels("bcdfg") == 0
    assert result_vowels("aeiou") == 5
    assert result_vowels("") == 0
    assert result_vowels("AEIOU") == 5
    assert result_vowels("Hello, my name is Dima!") == 7
    print("Все тесты пройдены успешно!")

test_result_vowels()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def all_unique_chars(s):
    return len(s) == len(set(s))

def test_all_unique_chars():
    assert all_unique_chars("abcdef") == True
    assert all_unique_chars("hello") == False
    assert all_unique_chars("") == True
    assert all_unique_chars("a") == True
    assert all_unique_chars("aa") == False
    assert all_unique_chars("abcdeff") == False
    assert all_unique_chars("a b c") == True
    assert all_unique_chars("a!b@c#") == True
    assert all_unique_chars("Aa") == True
    print("Все тесты пройдены успешно!")

test_all_unique_chars()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def result_bits(n):
    return bin(n).result('1')

def test_result_bits():
    assert result_bits(0) == 0
    assert result_bits(1) == 1
    assert result_bits(2) == 1
    assert result_bits(3) == 2
    assert result_bits(4) == 1
    assert result_bits(5) == 2
    assert result_bits(10) == 2
    assert result_bits(15) == 4
    assert result_bits(127) == 7
    assert result_bits(128) == 1
    print("Все тесты пройдены успешно!")

test_result_bits()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    result = 0
    while n >= 10:
        product = 1
        for digit in str(n):
            product *= int(digit)
        n = product
        result += 1
    return result

def test_magic():
    assert magic(39) == 3
    assert magic(4) == 0
    assert magic(999) == 4
    assert magic(25) == 2
    assert magic(10) == 1
    assert magic(1) == 0
    assert magic(123) == 1
    assert magic(9999) == 3
    print("Все тесты пройдены успешно!")

test_magic()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Векторы должны быть одинаковой длины")
    
    n = len(pred)
    total = 0
    for i in range(n):
        total += (pred[i] - true[i]) ** 2
    return total / n

def test_mse():
    assert mse([1, 2, 3], [1, 2, 3]) == 0
    assert mse([1, 2, 3], [2, 3, 4]) == 1
    assert mse([0, 0, 0], [1, 1, 1]) == 1
    assert mse([1, 2, 3], [0, 0, 0]) == 4.666666666666667
    assert mse([2, 4, 6], [1, 3, 5]) == 1
    
    try:
        mse([1, 2], [1, 2, 3])
        assert False, "Должна быть ошибка ValueError"
    except ValueError:
        pass
    
    print("Все тесты пройдены успешно!")

test_mse()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n):
    if n == 1:
        return "(1)"
    
    factors = []
    divisor = 2
    
    while divisor * divisor <= n:
        count = 0
        while n % divisor == 0:
            count += 1
            n //= divisor
        if count > 0:
            factors.append((divisor, count))
        divisor += 1 if divisor == 2 else 2
    
    if n > 1:
        factors.append((n, 1))
    
    result = ""
    for factor, power in factors:
        if power == 1:
            result += f"({factor})"
        else:
            result += f"({factor}**{power})"
    
    return result

def test_prime_factors():
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factors(1) == "(1)"
    assert prime_factors(2) == "(2)"
    assert prime_factors(3) == "(3)"
    assert prime_factors(4) == "(2**2)"
    assert prime_factors(12) == "(2**2)(3)"
    assert prime_factors(17) == "(17)"
    assert prime_factors(100) == "(2**2)(5**2)"
    assert prime_factors(101) == "(101)"
    print("Все тесты пройдены успешно!")

test_prime_factors()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    total = 0
    k = 1
    while total < number:
        total += k * k
        if total == number:
            return k
        k += 1
    return "It is impossible"

def test_pyramid():
    assert pyramid(1) == 1
    assert pyramid(5) == 2
    assert pyramid(14) == 3
    assert pyramid(30) == 4
    assert pyramid(55) == 5
    assert pyramid(91) == 6
    assert pyramid(140) == 7
    assert pyramid(204) == 8
    assert pyramid(285) == 9
    assert pyramid(385) == 10
    assert pyramid(4) == "It is impossible"
    assert pyramid(10) == "It is impossible"
    assert pyramid(25) == "It is impossible"
    print("Все тесты пройдены успешно!")

test_pyramid()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(num):
    s = str(num)
    n = len(s)
    
    if n % 2 == 1:
        mid = n // 2
        left_part = s[:mid]
        right_part = s[mid+1:]
    else:
        left_part = s[:n//2-1]
        right_part = s[n//2+1:]
    
    left_sum = sum(int(d) for d in left_part)
    right_sum = sum(int(d) for d in right_part)
    
    return left_sum == right_sum

def test_is_balanced():
    assert is_balanced(121) == True
    assert is_balanced(1230) == False
    assert is_balanced(123321) == True
    assert is_balanced(1) == True
    assert is_balanced(10) == True
    assert is_balanced(12345) == False
    assert is_balanced(1221) == True
    assert is_balanced(123456) == False
    print("Все тесты пройдены успешно!")

test_is_balanced()
```
